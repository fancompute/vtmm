```python
import numpy as np
import matplotlib.pylab as plt
from vtmm import tmm_rt
```

```python
pol = 'p'
c0 = 3e8
```

```python
lambdas = 1e-9 * np.linspace(100, 1000, 1000)
omegas = 2 * np.pi * c0 / lambdas
kxs = 0 * omegas / c0


n = np.array([2.0, 3.0 + 0.5j, 2.0, 1.0])
d = np.array([50e-9, 100e-9])
```

```python
rs, ts = tmm_rt(pol, omegas, kxs, n, d)
rs = rs.numpy()[0, :]
ts = ts.numpy()[0, :]
```

```python
plt.plot(omegas / c0 * n[1] * d[0], np.abs(rs)**2 / 2.0, label='$|R|^2$')
plt.plot(omegas / c0 * n[1] * d[0], np.abs(ts)**2, label='$|T|^2$')
plt.plot(omegas / c0 * n[1] * d[0], np.abs(rs)**2 / 2.0 + np.abs(ts)**2, label='$|R|^2 + |T|^2$')
plt.xlabel('$\omega~(rad/s)$')
# plt.ylabel('$|a|^2$')
plt.legend()
plt.title(f'd = {(d[0]*1e9):.2f} nm, n = {(n[1]):.2f}')
plt.savefig('transmission', dpi=300)
plt.show()
```
